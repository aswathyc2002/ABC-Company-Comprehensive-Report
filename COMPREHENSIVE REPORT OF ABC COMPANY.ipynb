```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing file URL from Github

dataset = "https://raw.githubusercontent.com/ajaythomas200/ABC_Company_Comprehensive_Report_Python/refs/heads/main/ABC%20Company%20Dataset.csv.csv"
```

```python
data = pd.read_csv(dataset)
data
```

```python
print(data.info())
```

```python
data.isnull().sum()
```

```python
# Filling the null salary  columns with mean value

data['Salary'].fillna(data['Salary'].mean(),inplace = True)
```

```python
# Filling the null college names to unknown

data['College'].fillna('Unknown',inplace = True)
```

```python
# Removing Duplicates entry

data.drop_duplicates(inplace = True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
# Correcting the height column 

data['Height'] = np.random.randint(150,181, size = len(data))
```

```python
data
```

# Task 1

```python
#1. Employees distribution across each team

employees_distribution = data['Team'].value_counts()
print("Employees distribution across each team:\n")
print(employees_distribution)
```

```python
# Calculating percentage split relative to the total number of employees

team_percentage = (employees_distribution / len(data)) * 100
print("Percentage splitting with respect to the total employees:\n")
print(team_percentage)
```

```python
# Bar plot for showing the employees distribution across the team
plt.figure(figsize=(12, 6))
x =employees_distribution.index
y=employees_distribution.values
sns.barplot(x=employees_distribution.index, y=employees_distribution.values, palette="viridis",)
plt.xlabel("Teams",fontweight='bold')
plt.ylabel("Number of Employees",fontweight='bold')
plt.title("Employee Distribution Across Teams",fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

# Task 2

```python
#2. Segragate employees based on their positions with in the company

positions_distribution = data['Position'].value_counts()
print("Employee Distribution Across Positions:\n")
print(positions_distribution)
```

```python
positions_percentage = (positions_distribution / len(data)) * 100
print("Percentage Split Relative to Total Employees:\n")
print(positions_percentage)
```

```python
# Creating a pie chart for employee distribution across positions

plt.figure(figsize=(7, 7))
plt.pie(positions_distribution, labels=positions_distribution.index, autopct='%1.1f%%', 
        colors=sns.color_palette("viridis", len(positions_distribution)), startangle=0)
plt.title("Employee Distribution Across Positions (Percentage)",fontweight='bold')
plt.show()
```

# Task 3

```python
#3. Identifying the predominant age group among employees

age_bins = [18, 25, 30, 35, 40, 45, 50, 60]  # Age ranges
age_labels = ["18-24", "25-29", "30-34", "35-39", "40-44", "45-49", "50+"]

data['Age Group'] = pd.cut(data['Age'], bins = age_bins, labels = age_labels, right=False)
age_group_distribution = data['Age Group'].value_counts().sort_index()

print("Employee Distribution Across Age Groups:\n")
print(age_group_distribution)
```

```python
# Creating  a bar plot for showing age group distribution

plt.figure(figsize=(8, 6))
sns.barplot(x = age_group_distribution.index, y = data["Age Group"].value_counts().sort_index().values, palette="viridis")
plt.xlabel('Age Group',fontweight='bold')
plt.ylabel('Number of Employees',fontweight='bold')
plt.title('Distribution of Employees by Age Group',fontweight='bold')
for index, value in enumerate(data['Age Group'].value_counts().sort_index().values):
    plt.text(index, value + 2, str(value), ha='center')
plt.grid(axis='y', linestyle='--', color ='black', alpha=0.7, zorder=0)
plt.gca().set_axisbelow(True)
plt.show()
```

# Task 4

```python
#4.  Calculate total salary expenditure by team and position
team_salary_expenditure = data.groupby('Team')['Salary'].sum().sort_values(ascending=False)
position_salary_expenditure = data.groupby('Position')['Salary'].sum().sort_values(ascending=False)

highest_team = team_salary_expenditure.idxmax()
highest_position = position_salary_expenditure.idxmax()
highest_team_salary = team_salary_expenditure.max()
highest_position_salary = position_salary_expenditure.max()

print('Highest Team :',highest_team)
print('Highest Team  Salary  :',highest_team_salary)
print('Highest Position :',highest_position)
print('Highest Position Salary :',highest_position_salary)
```

```python
# Horizontal bar plot for showing Total salary expenditure by team

plt.figure(figsize=(12, 8))
sns.barplot(y=team_salary_expenditure.index, x=team_salary_expenditure.values, palette="viridis")
plt.xlabel('Total Salary Expenditure ($)',fontweight='bold')
plt.ylabel('Team', fontweight='bold')
plt.title('Total Salary Expenditure by Team',fontweight='bold')
for index, value in enumerate(team_salary_expenditure.values):
    plt.text(value + 2, index, f"${value:,.0f}", va='center')
plt.show()
```

```python
# Vertical bar plot for Salary Expenditure by Position
plt.figure(figsize=(10, 5))
sns.barplot(x=position_salary_expenditure.index, y=position_salary_expenditure.values, palette='viridis')
plt.ylabel('Total Salary Expenditure ($)',fontweight='bold')
plt.xlabel('Position',fontweight='bold')
plt.title('Total Salary Expenditure by Position',fontweight='bold')
plt.xticks(rotation=90)
plt.ylim(0, position_salary_expenditure.max() * 1.1)
for index, value in enumerate(position_salary_expenditure.values):
    plt.text(index, value + (value * 0.01), f"${value:,.0f}", ha='center')
plt.show()
```

# Task 5

```python
# Finding the correlation between age and salary
correlation = data["Age"].corr(data["Salary"])
print(f"Correlation between Age and Salary: {correlation:.2f}")
```

```python
# scatter plot for showing correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data["Age"], y=data["Salary"], hue=data["Age"], palette="viridis")
plt.xlabel("Age", fontweight="bold")
plt.ylabel("Salary ($)", fontweight="bold")
plt.title(f"Age vs. Salary (Correlation: {correlation:.2f})", fontsize=14, fontweight="bold")
plt.grid()
plt.show()
```

# **Data Story: Insights from ABC Company Employee Analysis**

# **Overview**
This analysis aimed to explore ABC Company's employee dataset by examining various aspects such as team distribution, employee positions, age groups, salary expenditures, and correlations between age and salary. After preprocessing the dataset, key insights were derived using statistical analysis and visualizations.
The dataset consists of 458 employees and 9 attributes, providing key details about employees across various teams in ABC Company. 

# **Dataset**
The dataset includes the following columns:

- **Employee ID** – Unique identifier for each employee  
- **Name** – Full name of the employee  
- **Age** – Employee’s age  
- **Height** – Employee’s height (corrected with random values between 150 cm and 180 cm for consistency)  
- **Weight** – Employee’s weight  
- **Team** – Department or team the employee belongs to  
- **Position** – Job title of the employee  
- **Salary** – Employee's annual salary in USD  
- **College** – Educational institution attended by the employee  

## **Key Insights and Analysis**

### **1. Employee Distribution Across Teams**
- The distribution of employees across different teams was analyzed.
- The orleans Pelicans have the highest number of employees.
- A bar chart effectively visualized the proportion of employees per team.

### **2. Employee Segregation Based on Position**
- The most common position is SG (Shooting Guard), accounting for 22% of all employees.
- A pie chart provided a visual representation of job role distribution.

### **3. Predominant Age Group Among Employees**
- Employees were grouped into age brackets (e.g., 18-24, 25-29, 30-34, etc.).
- The most common age group is 25-29 years, with 182 employees in this range.
- A bar chart was used to represent the age distribution.

### **4. Highest Salary Expenditure by Team and Position**
- The team with the highest salary expenditure is Cleveland, indicating the highest salary investments.
- The position with the highest salary expenditure is C Group, suggesting key roles with high compensation.
- Bar charts (horizontal and vertical) were used for clear visualization.

### **5. Correlation Between Age and Salary**
- The correlation coefficient between Age and Salary is 0.21, indicating a weak positive correlation.
- This suggests that age slightly influences salary but is not the main determining factor.
- A scatter plot was used to examine the relationship.

This analysis provides valuable insights into employee distribution, salaries, and workforce trends.
